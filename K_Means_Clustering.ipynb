```python
# CLUSTERING =====>>
# K - MEANS CLUSTERING 
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading iris dataset into iris variable
iris = load_iris()
Data_Iris = iris.data
# iris
```

```python
# Importing KMeans algo
from sklearn.cluster import KMeans
# # setting K value to 3 i.e. n_clusters = 3
KMNS = KMeans(n_clusters=3)
# fitting iris data set into KMNS var.
KMNS.fit(Data_Iris)
# predicting the target labels
labels = KMNS.predict(Data_Iris)
labels
#  getting the centres of the clusters
cnt = KMNS.cluster_centers_
cnt
```

```python
# plotting the KMEANS GRAPH USING PETAL LENGTH AND PETAL WIDTH
plt.scatter(Data_Iris[:,2],Data_Iris[:,3], c = labels)
plt.scatter(cnt[:,2],cnt[:,3],color='red',marker='o',s=120)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()

```

```python
# Inertia values List
K_Inertias = [] 
# # setting K value to 3 i.e. n_clusters = 3
# KMNS = KMeans(n_clusters=3)
for i in range(1,10):
  KMNS = KMeans(n_clusters=i, random_state=44)
  # fitting iris data set into KMNS var.
  KMNS.fit(Data_Iris)
  # predicting the target labels
  labels = KMNS.predict(Data_Iris)
  labels

  #  getting the centres of the clusters
  cnt = KMNS.cluster_centers_
  cnt
  # appending intertia value into list
  K_Inertias.append(KMNS.inertia_)
K_Inertias
# plotting intertia graph
plt.plot(range(1,10), K_Inertias, color='green', marker='o')
plt.xlabel('number of K')
plt.ylabel('Inertia')
plt.show()
```
